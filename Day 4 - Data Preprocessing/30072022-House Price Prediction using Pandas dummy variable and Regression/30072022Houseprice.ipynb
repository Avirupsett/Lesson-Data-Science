```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/sample_data/homeprice - Sheet1.csv")
df.head()
```

```python
df_dummies=pd.get_dummies(df['town'])
df_dummies
```

```python
all_val=pd.concat([df,df_dummies],axis=1)
all_val
```

```python
x=all_val.drop(columns=['town','price','monroe township'])
x
```

```python
y=all_val['price']
y
```

```python
reg_model= linear_model.LinearRegression()
reg_model.fit(x,y)
```

```python
reg_model.predict([[3100,0,0]])
```

```python
reg_model.score(x,y)
```
