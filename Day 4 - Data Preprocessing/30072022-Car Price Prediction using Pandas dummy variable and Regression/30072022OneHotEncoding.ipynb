```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/sample_data/carprice - Sheet1.csv")
df.head()
```

```python
df_dummies=pd.get_dummies(df['Car Model'])
df_dummies
```

```python
all_val=pd.concat([df,df_dummies],axis=1)
all_val
```

```python
x=all_val.drop(columns=['Car Model','Sell Price($)','BMW X5'])
x
```

```python
y=all_val['Sell Price($)']
y
```

```python
reg_model= linear_model.LinearRegression()
reg_model.fit(x,y)
```

```python
reg_model.predict([[55500,2,1,0]])
```

```python
reg_model.score(x,y)
```
